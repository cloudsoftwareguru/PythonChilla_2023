Exploratory Data Analysis is a method used to analyse and summarize datasets. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
# print(df.head())
# print(df.tail())
df
```

```python
sns.heatmap(df.isnull(), yticklabels=True, cbar=True, cmap='viridis')
```

```python
df.drop(columns="deck", inplace=True)
```

```python
df.shape
```

```python
#visualize and explore
sns.countplot(data=df, x="survived")
```

```python
sns.barplot(data=df, x="survived", y="age", hue="who")
```

```python
sns.barplot(data=df, x="survived", y="fare", hue="who")
```

```python
sns.countplot(x="pclass", data=df)
```

```python
sns.pairplot(df, hue='survived')
```

```python
#stacked barchart/plot

df.groupby(['pclass', 'sex'])
```
